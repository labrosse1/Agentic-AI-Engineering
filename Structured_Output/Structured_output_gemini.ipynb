```python
import json
from pydantic import BaseModel, Field
from google import genai
from google.genai import types
from google.colab import userdata

api_key = userdata.get('GOOGLE_API_KEY')

client = genai.Client(api_key=api_key)

MODEL_ID = "gemini-2.5-flash"
```

```python
class DocumentMetadata(BaseModel):
    """Class to hold structured metadata for a document passed to the LLM."""

    exercise_name: str = Field(description="The name of the exercice.")
    instructions: list[str] = Field(description="Explanation on how to do the exercise.")
    targeted_muscles: list[str] = Field(description="Muscles targeted by this exercise (e.g, Brachial).")
    why: str = Field(description="Which part of the body is targeted by this exercise.")
```

```python
with open('/content/page1.png', 'rb') as f:
    image = f.read()

#Pass an image with new google-genai SDK

document = types.Part.from_bytes(
    data=image,
    mime_type="image/png"
)
```

```python
prompt = f"""
Analyze the following document and extract its metadata.
If the text is not in english, keep the output in the initial language present on the document.

Here is the document:
<document>
{document}
</document>
"""
```

```python
gemini_config = types.GenerateContentConfig(response_mime_type="application/json", response_schema=DocumentMetadata)

response = client.models.generate_content(model=MODEL_ID, contents=[document, prompt], config=gemini_config)
response
```
